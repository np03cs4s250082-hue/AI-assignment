```python
import numpy as np
```

```python
# Problem 1:
print("Problem 1: Array Creation")
empty_array = np.empty((2, 2))
print("1. Empty array (2x2):\n", empty_array)
ones_array = np.ones((4, 2))
print("\n2. All ones array (4x2):\n", ones_array)
full_array = np.full((3, 3), 7)
print("\n3. Full array (3x3) with 7:\n", full_array)
ref_array = np.array([[1, 2], [3, 4]])
zeros_like_array = np.zeros_like(ref_array)
print("\n4. Zeros like ref_array:\n", zeros_like_array)
ones_like_array = np.ones_like(ref_array)
print("\n5. Ones like ref_array:\n", ones_like_array)
new_list = [1, 2, 3, 4]
converted_array = np.array(new_list)
print("\n6. Converted list to array:\n", converted_array)
```

```python
# Problem 2:
print("\nProblem 2: Array Manipulation")
range_array = np.arange(10, 50)
print("1. Array from 10 to 49:\n", range_array)
matrix_3x3 = np.arange(9).reshape(3, 3)
print("\n2. 3x3 matrix from 0 to 8:\n", matrix_3x3)
identity_matrix = np.eye(3)
print("\n3. Identity matrix (3x3):\n", identity_matrix)
random_array = np.random.random(30)
print("\n4. Random array (size 30):\n", random_array)
print("Mean of array:", random_array.mean())
rand_10x10 = np.random.random((10, 10))
print("\n5. Random 10x10 array:\n", rand_10x10)
print("Min:", rand_10x10.min(), "Max:", rand_10x10.max())
zero_array = np.zeros(10)
zero_array[4] = 1
print("\n6. Zero array with 5th element = 1:\n", zero_array)
arr = [1, 2, 0, 0, 4, 0]
reversed_arr = arr[::-1]
print("\n7. Reversed array:\n", reversed_arr)
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n8. 2D array with 1 on border and 0 inside:\n", border_array)
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n9. 8x8 checkerboard pattern:\n", checkerboard)
```

```python
# Problem 3:
print("\nProblem 3: Array Operations")
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])
print("x:\n", x)
print("y:\n", y)
print("v:", v)
print("w:", w)
print("\n1. x + y:\n", x + y)
print("2. x - y:\n", x - y)
print("3. x * 3:\n", x * 3)
print("4. Square of x:\n", x ** 2)
print("\n5. Dot products:")
print("v . w =", np.dot(v, w))
print("x . v =\n", np.dot(x, v))
print("x . y =\n", np.dot(x, y))
print("\n6. Concatenate x and y along rows:\n", np.concatenate((x, y), axis=0))
print("Concatenate v and w along columns:\n", np.vstack((v, w)))
print("\n7. Concatenate x and v:")
try:
    result = np.concatenate((x, v), axis=0)
    print(result)
except ValueError as e:
    print("Error:", e)
```

```python
# Problem 4:
print("\nProblem 4: Matrix Operations")
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("1. A * A_inv:\n", identity_check)
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("BA:\n", BA)
print("AB == BA?", np.allclose(AB, BA))
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)^T:\n", AB_T)
print("B^T * A^T:\n", BT_AT)
print("(AB)^T == B^T * A^T?", np.allclose(AB_T, BT_AT))
```

```python
print("\nSolving Linear Equations")
A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_sys = np.array([-1, -3, 9])
A_inv_sys = np.linalg.inv(A_sys)
X = np.dot(A_inv_sys, B_sys)
print("Solution [x, y, z]:", X)
```

```python
import time
size = 1_000_000
matrix_size = 1000
print("Experiment: How Fast is NumPy?\n")
```

```python
# 1. Element-wise Addition
print("1. Element-wise Addition")
list1 = list(range(size))
list2 = list(range(size))
start = time.time()
list_add = [a + b for a, b in zip(list1, list2)]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")
arr1 = np.arange(size)
arr2 = np.arange(size)
start = time.time()
np_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds\n")
```

```python
# 2. Element-wise Multiplication
print("2. Element-wise Multiplication")
start = time.time()
list_mul = [a * b for a, b in zip(list1, list2)]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")
start = time.time()
np_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds\n")
```

```python
# 3. Dot Product
print("3. Dot Product")
start = time.time()
list_dot = sum(a * b for a, b in zip(list1, list2))
end = time.time()
print("Python List Dot Product Time:", end - start, "seconds")
start = time.time()
np_dot = np.dot(arr1, arr2)
end = time.time()
print("NumPy Array Dot Product Time:", end - start, "seconds\n")
```

```python
# 4. Matrix Multiplication
print("4. Matrix Multiplication")
list_matrix1 = [[i for i in range(matrix_size)] for _ in range(matrix_size)]
list_matrix2 = [[i for i in range(matrix_size)] for _ in range(matrix_size)]
start = time.time()
result_matrix = [[sum(a * b for a, b in zip(row, col)) for col in zip(*list_matrix2)] for row in list_matrix1]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")

np_matrix1 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
np_matrix2 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)

start = time.time()
np_result_matrix = np.dot(np_matrix1, np_matrix2)
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
